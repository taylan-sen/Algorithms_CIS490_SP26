```python
# SEND A UDP PACKET
import socket  # main module for talking to transport lyr

# create a socket obj
s = socket.socket(type=socket.SOCK_DGRAM)

# prepare msg
msg = 'Ty says listen to the AI networking songs'

# convert 1's and zeros encode
josh = msg.encode()

# send packet
s.sendto(josh, ('35.192.97.212',12000))

s.close()
print('program complete')
```

```python
# SEND A TCP PACKET
import socket  # main module for talking to transport lyr

# create a socket obj
s = socket.socket()

# prepare msg
msg = 'TCP MAKES A CONNECTION and is RELIABLE AND SLOW'

# convert 1's and zeros encode
msg_bytes = msg.encode()

# open a connection
s.connect(('35.192.97.212',12001))

# send packet
s.sendall(msg_bytes)

s.close()
print('program complete')
```
